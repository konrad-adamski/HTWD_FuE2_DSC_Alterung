### b) Sortieren

```python
def sort_columns(df, last_columnname='Aging Level'):
    """
    Sortiert die Spalten eines DataFrames in folgender Reihenfolge:
    1. Spalten, die nicht mit 'T' oder 'Peak Signal' beginnen und nicht die letzte Spalte sind.
    2. Spalten, die mit 'T' beginnen.
    3. Spalten, die mit 'Peak Signal' beginnen.
    4. Die letzte Spalte (standardmäßig 'Aging Level').

    Parameters:
        df (pd.DataFrame): Der zu sortierende DataFrame.
        last_columnname (str): Name der letzten Spalte (Standard: 'Aging Level').

    Returns:
        pd.DataFrame: DataFrame mit neu sortierten Spalten.
    """
    columns = df.columns

    # Listen für die Sortierung
    first_columns = [col for col in columns if not col.startswith(('T', 'Peak Signal')) and col != last_columnname]
    t_columns = [col for col in columns if col.startswith('T')]
    peak_columns = [col for col in columns if col.startswith('Peak Signal')]
    last_column = [last_columnname] if last_columnname in columns else []

    # Neu sortieren: erst 'first_columns', dann 't_columns', dann 'peak_columns', dann 'last_column'
    sorted_columns = first_columns + t_columns + peak_columns + last_column

    # DataFrame mit neu sortierten Spalten zurückgeben
    return df[sorted_columns]
```

```python
# Aufruf
df_sorted_heating = sort_columns(df_merged_heating)
df_sorted_heating
```

```python
df_sorted_heating.to_csv("data/02_Data_Heating_merged.csv", index = False)
```

## II) Dateien umbennen 

> damit sie mit Werten aus der "Name of Original DSC Measurement"-Spalte übereinstimmen

```python
import os
import re
import shutil

def rename_files_in_folder(input_folder, output_folder):
    
    # Zähler für bearbeitete Dateien
    processed_count = 0
    
    # Eingabe- und Ausgabeordner überprüfen
    if not os.path.exists(input_folder):
        print(f"Der Eingabeordner '{input_folder}' existiert nicht. Bitte überprüfe den Pfad.")
        return
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)  # Ausgabeordner erstellen, falls nicht vorhanden

    # Überprüfen, ob Dateien mit "ExpDat_" vorhanden sind
    files = [f for f in os.listdir(input_folder) if f.startswith("ExpDat_")]
    if not files:
        print("Keine passenden Dateien gefunden.")
        return

    # Dateien umbenennen und in den Ausgabeordner kopieren
    for filename in files:
        if os.path.isfile(os.path.join(input_folder, filename)):
            # Präfix "ExpDat_" entfernen und Suffix "-<Zahl>" direkt vor der Erweiterung entfernen
            new_filename = re.sub(r'^ExpDat_|-\d+(?=\.\w+$)', '', filename)
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, new_filename)
            shutil.copy(input_path, output_path)  # Datei kopieren
            processed_count += 1
            
    # Ergebnis ausgeben
    print(f"{processed_count} Dateien wurden bearbeitet und kopiert.")
    
    
# Aufruf
rename_files_in_folder("original_data/all", "Textfiles")
```
